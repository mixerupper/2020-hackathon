```python
pageview["en"] = 0
pageview["es"] = 0

en_idx = pageview[pageview["URL_PATH"].str.startswith("/en")].index
es_idx = pageview[pageview["URL_PATH"].str.startswith("/es")].index

pageview.loc[en_idx, "en"] = 1
pageview.loc[es_idx, "es"] = 1

pageview["URL_PATH"] = pageview["URL_PATH"].str.replace("/en", "")
pageview["URL_PATH"] = pageview["URL_PATH"].str.replace("/es", "")

pageview_clean = pageview[~(pageview["URL_PATH"] == "")].copy()
```
