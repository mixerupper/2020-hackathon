```python
pageview["en"] = 0
pageview["es"] = 0

en_idx = pageview[pageview["URL_PATH"].str.startswith("/en")].index
es_idx = pageview[pageview["URL_PATH"].str.startswith("/es")].index

pageview.loc[en_idx, "en"] = 1
pageview.loc[es_idx, "es"] = 1

pageview["URL_PATH"] = pageview["URL_PATH"].str.replace("/en", "")
pageview["URL_PATH"] = pageview["URL_PATH"].str.replace("/es", "")

pageview_clean = pageview[~(pageview["URL_PATH"] == "")].copy()
```

```python
temp = pageview_clean.join(content.set_index("url"), on="URL_PATH", how="inner")
temp.reset_index(drop=True, inplace=True)
```

```python
grouped_by_user = temp.groupby("USER_ID")
```

```python
urls = temp["URL_PATH"].unique()
users = temp["USER_ID"].unique()

initial_values = np.zeros((users.shape[0], urls.shape[0]), dtype=np.uint8)
matrix = pd.DataFrame(initial_values, columns=urls, index=users)
```

```python
groups = grouped_by_user.groups.keys()
```
